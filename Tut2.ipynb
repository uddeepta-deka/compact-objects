## Deriving homology relations

The equilibrium stellar structure equations (ref. tutorial 1) are:
\begin{align}
    \frac{dr}{dM_r} &= \frac{1}{\rho\, 4\pi r^2}\label{eqm1}\\
    \frac{dP}{dM_r} & = -\frac{GM_r}{4\pi r^4}\label{eqm2}\\
    \frac{dT}{dM_r} & = -\frac{3\kappa}{64\pi^2 ac}\frac{L_r}{T^3r^4}\label{eqm3}\\
    \frac{dL_r}{dM_r} & = \epsilon\label{eqm4}
\end{align}
where $M_r\equiv M(r)$ and $L_r \equiv L(r)$. 

Now, two stars of masses $M_1$, $M_2$ and radii $R_1$, $R_2$, respectively, are said to be homologous if their relative mass coordinate $m\equiv M_r/M$ and their relative radii $x\equiv r/R$ satisfy the following:
\begin{align}
    m &= \frac{M_{r1}}{M_1} = \frac{M_{r2}}{M_2}\label{hc1}\\
    x &=  \frac{r_1(m)}{R_1} = \frac{r_2(m)}{R_2}\label{hc2}
\end{align}
for all $m$.

### Density
For star 1, eqn.\eqref{eqm1} can be written as:
\begin{equation}
    \frac{dr_1}{dM_{r1}}= \frac{1}{\rho_1\, 4\pi r_1^2}\implies\frac{dr_1}{dm} = \frac{M_1}{\rho_1\, 4\pi r_1^2}
\end{equation}
Since the stars are homologous, we use eqn\eqref{hc2} to substitute $r_1 = r_2(R_1/R_2)$ in the above equation to get:
\begin{equation}
    \frac{R_1dr_2}{R_2dm} = \frac{M_1R_2^2}{\rho_1\, 4\pi r_2^2R_1^2}\implies \frac{dr_2}{dm} = \frac{M_2}{\rho_2\, 4\pi r_2^2}\left(\frac{\rho_2M_1R_2^3}{\rho_1M_2R_1^3}\right)
\end{equation}

Comparing this with eqn\eqref{eqm1} for star 2, we observe that the term inside the bracket in the above equation should equate to 1. Thus, 
\begin{equation}
    \frac{\rho_2M_1R_2^3}{\rho_1M_2R_1^3}=1\implies \rho(m)\propto MR^{-3}\label{homoeqn1}
\end{equation}

Additionally, we note that this relation must hold at any homologous mass shell, and therefore at the centre of each star. Also, $MR^{-3}\propto \bar{\rho}$ (the average stellar density). Thus, 
\begin{equation}
    \rho(m)\propto\rho_c\propto \bar{\rho}\propto MR^{-3}
\end{equation}

### Pressure
For star 1, eqn.\eqref{eqm2} can be written as:
\begin{equation}
    \frac{dP_1}{dM_{r1}} = -\frac{GM_{r1}}{4\pi r_1^4}\implies\frac{dP_1}{dm} = -\frac{GM_1^2m}{4\pi r_1^4}
\end{equation}
Substituting $r_1 = r_2(R_1/R_2)$ in the above equation, we get:
\begin{equation}
    \frac{dP_1}{dm} = -\frac{GM_1^2mR_2^4}{4\pi r_2^4R_1^4} = -\frac{GM_2^2m}{4\pi r_2^4}\left(\frac{M_1^2R_2^4}{M_2^2R_1^4}\right) = \frac{dP_2}{dm}\left(\frac{M_1^2R_2^4}{M_2^2R_1^4}\right)
\end{equation}

Upon integration w.r.t. $m$, the above equation takes the form:
\begin{equation}
    P_1(m) = P_2(m)\left(\frac{M_1^2R_2^4}{M_2^2R_1^4}\right)+\text{const.}
\end{equation}

The constant of integration should equate to zero because, for both stars, as $m\to 1$ at the stellar surface, $P\to 0$. Thus, 
\begin{equation}
    \frac{P_1(m)}{P_2(m)} = \frac{M_1^2R_2^4}{M_2^2R_1^4}\label{homoeqn2}
\end{equation}

This is true for all homologous mass shell, which must include the centre. Hence, for all $m$,
\begin{equation}
    P(m)\propto P_c\propto\frac{M^2}{R^4}
\end{equation}

### Temperature
Let us consider ideal gas equation of state:
\begin{equation}
    P=\frac{\mathcal{R}}{\mu}\rho T
\end{equation}
where $\mathcal{R}$ is the gas constant and $\mu$ is the mean molecular weight. Further assuming, $\mu$ is constant for both stars, we can obtain the following from eqns.\eqref{homoeqn1},\eqref{homoeqn2}:
\begin{equation}
    \frac{P_1(m)}{P_2(m)} = \frac{\mu_2}{\mu_1}\frac{\rho_1(m)T_1(m)}{\rho_2(m)T_2(m)}\implies\frac{T_2(m)}{T_1(m)} = \frac{\mu_2M_2R_1}{\mu_1M_1R_2}\label{homoeqn3}
\end{equation}

Thus, 
\begin{equation}
    T(m)\propto T_c\propto\mu\frac{M}{R}
\end{equation}

### Luminosity
Let us consider constant opacity along with previous assumptions of homogeneity and ideal gas. Eq.\eqref{eqm3} for star 1 gives,
\begin{equation}
    \frac{dT_1}{dM_{r1}} = -\frac{3}{64\pi^2 ac}\frac{\kappa_1L_{r1}}{T_1^3r_1^4}
\end{equation}

Writing $L_r = Ll$ and using eq.\eqref{homoeqn3}, we get:
\begin{align}
    \frac{dT_1(m)}{dm} &= -\frac{3l}{64\pi^2 ac}\frac{\kappa_1L_1M_1}{T_1^3r_1^4} = -\frac{3l}{64\pi^2 ac} \frac{\kappa_1L_1R_1^3}{\mu_1^3M_1^2r_1^4}\\
    \implies\frac{dT_1(m)}{dm} &= -\frac{3l}{64\pi^2 ac}\frac{\kappa_2L_2R_2^3}{\mu_2^3M_2^2r_2^4}\left(\frac{\kappa_1\mu_2^3M_2^2R_2L_1}{\kappa_2\mu_1^3M_1^2R_1L_2}\right)=\frac{dT_2(m)}{dm}\left(\frac{\kappa_1\mu_2^3M_2^2R_2L_1}{\kappa_2\mu_1^3M_1^2R_1L_2}\right)
\end{align}

Then a similar reasoning as was held for pressure gives,
\begin{align}
    \frac{T_1(m)}{T_2(m)}&=\frac{\kappa_1\mu_2^3M_2^2R_2L_1}{\kappa_2\mu_1^3M_1^2R_1L_2}\\
    \implies \frac{L_1(m)}{L_2(m)} = \frac{\kappa_2\mu_1^4M_1^3}{\kappa_1\mu_2^4M_2^3}
\end{align}

Thus,
\begin{equation}
    L\propto L_s\propto\frac{1}{\kappa}\mu^4M^3
\end{equation}

where $L_s$ denotes the surface luminosity of a star. 
These homologous relations are consistent with the scaling we obtained in prob(4) of previous tutorial.

## $P_c-\rho_c$ relation for a star in quasi equilibrium

From the homologous relations we have, 
\begin{align}
    \rho_c\propto & \frac{M}{R^3}\implies R\propto\left(\frac{M}{\rho_c}\right)^{1/3}\\
    P_c\propto & \frac{M^2}{R^4}\implies P_c\propto\frac{M^2}{(M/\rho_c)^{4/3}}\implies P_c\propto M^{2/3}\rho_c^{4/3}
\end{align}

Doing a quick dimension analysis gives,
\begin{equation}
    P_c = C\cdot GM^{2/3}\rho_c^{4/3}
\end{equation}
where $C$ is a dimensionless constant

## Slowly contracting star in quasi-hydrostatic equilibrium

\begin{equation}
P_c = \frac{\mathcal{R}}{\mu}\rho_cT_c+K\left(\frac{\rho_c}{\mu_e}\right)^\gamma\implies T_c = \frac{\mu}{\mathcal{R}\rho_c}\left[CGM^{2/3}\rho_c^{4/3}-K\left(\frac{\rho_c}{\mu_e}\right)^\gamma\right]
\end{equation}

```python
#defining the constants

X = 0.75
mu = 0.5*pow(X,-0.57)
Rgas = 8.3e7

C = 0.5
G = 6.67e-8
Msun = 2e33

mue = 1
K = 1e13
```

```python
Mstar = np.array([.5 , 1., 10., 100.])*Msun #input masses of stars
```

```python
rhoc = np.logspace(-5,10) #specify range of rho_c
tcM = []
gamma = np.linspace(4./3.,5./3.,10)
for M in Mstar:
    tc=[]
    for g in gamma:
        temp = []
        for i in range(len(rhoc)):
            temp.append((mu/(Rgas*rhoc[i]))*(C*G*pow(M,2./3.)*pow(rhoc[i],4./3.)-K*pow(rhoc[i]/mue,g)))
        tc.append(temp)
    tcM.append(tc)
```

Here I plot variation of $T_c$ with $\rho_c$ for different masses of stars

```python
Ms = Mstar/Msun
plt.figure(figsize = (10,30))
plt.suptitle(r'Variation of $T_c$ with $\rho_c$',size = 24,y=.92)
plt.subplots_adjust(hspace=0.4)
for j in range(len(Mstar)):
    plt.subplot(len(Mstar),1,j+1)
    for i in range(len(gamma)):
        plt.loglog(rhoc,tcM[j][i],'o-',label=r'$\gamma=$%3.2f'%gamma[i])
    plt.title(r'Mass of star = %3.1f$M_\odot$'%Ms[j],size = 20)
    plt.xlabel(r'$\rho_c$')
    plt.ylabel(r'$T_c$')
plt.legend(bbox_to_anchor=(1.5, 3.))
plt.show()
```

## Deriving expression for maximum central temperature reached by a star of mass $M$

We have,
\begin{equation}
    T_c = \frac{\mu}{\mathcal{R}\rho_c}\left[CGM^{2/3}\rho_c^{4/3}-K\left(\frac{\rho_c}{\mu_e}\right)^\gamma\right]
\end{equation}

At the maximum value of $T_c$ we have
\begin{align}
    \frac{dT_c}{d\rho_c} & =\frac{\mu}{3\mathcal{R}\rho_c^2}\left[C G M^{2/3}\rho_c^{4/3}-3K(\gamma-1)\left(\frac{\rho_c}{\mu_e}\right)^\gamma\right]=0 \\
    \implies \rho_c &= \left(\frac{3K(\gamma-1)}{CG\mu_e^{\gamma}}\right)^{\frac{3}{4-3\gamma}}M^{-2/3}
\end{align}

Thus,
\begin{align}
    (T_c)_{max} &= \frac{\mu}{\mathcal{R}}\left(\frac{CG}{3}\right)^{\frac{3(\gamma-1)}{3\gamma-4}}K^{\frac{-1}{3\gamma-4}}\left(3 (\gamma-1)^\frac{-4}{3\gamma-4} - (\gamma-1)^\frac{-3\gamma}{3\gamma-4}\right)(\gamma-1)^{\frac{3}{3\gamma-4}}\mu_e^{\frac{\gamma}{3\gamma-4}} M^{\frac{2(\gamma-1)}{3\gamma-4}}\\
    \implies (T_c)_{max} & \propto M^{\frac{2(\gamma-1)}{3\gamma-4}}
\end{align}

Hence, the maximum central temperature reached by a star of mass $M$ increases with $M$ for a fixed $\gamma$.
